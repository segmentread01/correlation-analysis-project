# Partie 2.1 : qualité de données - valeurs abérrantes
En suite, je contrôle la qualité de données. Commençons par les valeurs abérrantes.

```python
df_perf_copy = df_perf.copy()
```

```python
def outliers_val(x):
    q1 = np.quantile(x, 0.25)
    q3 = np.quantile(x, 0.75)
    iqr = q3 - q1
    fence_high = q3 + 1.5*iqr
    
    col_val = x.loc[(x > fence_high)| x < 0] # position outliers (nom de colonne et valeur abérrante)
    #index_ligne = np.where(x >= (q3 + 1.5*iqr)) # position des outliers (seulement index ligne)
    
    return col_val #x.loc[x > fence_high]
```

```python
def detect_out(s):
        q1 = np.nanquantile(s,0.25)
        q3 = np.nanquantile(s,0.75)
        iqr = q3 - q1
        fence_high = q3 + 1.5*iqr
        s_val = s.loc[(s> fence_high)|(s< 0)]
        print(fence_high) 
        return s_val
```

```python
# Je reprends les colonnes de type float afin d'appliquer la fonction personnalisée (exclure nutrition_grade et nutrition_score)
perf_outlier = df_perf_copy.iloc[:,4:-1]
perf_outlier.dropna(axis=1, how='all', inplace=True)
perf_outlier.columns
```

Identifier et quantifier les valeurs aberrantes:

```python
for x, y in perf_outlier.iteritems():
    out_value = detect_out(y)
    print(x + ' ' + str(out_value.size))
    print(out_value[0:10])
```

**EXTRACTION FLAG DE CATEGORIES**

```python
import re
```

```python
cat_col = df["categories_fr"].unique()
```

```python
cat_processed = [re.split(',', str(cat)) for cat in cat_col]
```

```python
flatten = lambda l : [el for sublist in l for el in sublist]
```

```python
cat_flattened = flatten(cat_processed)
```

```python
count_dic = {}
for cat in cat_flattened:
    if cat not in count_dic:
        count_dic[cat] = 1
    else:
        count_dic[cat] = count_dic[cat] + 1
```

```python
sorted_cats = sorted([(key, value) for key, value in count_dic.items()], key= lambda el : el[1], reverse=True)
```

```python
substr_1 = ["végétaux", "végétal", "végétale","vegetal", "fruit", "légume"]
substr_2 = ["snacks", "dessert", "biscuit", "gâteaux", "sucre"]
substr_3 = ["epiceries", "tartiner", "condiment"]
substr_4 = ["la mer", "poisson", "fruit de mer"]

# Construire une liste de mots clé catégories
corpus_cat = [substr_1, ["plat préparé"], ["laitier"], 
            substr_2, ["céréales"], ["viande"], ["charcuterie"], ["fromages"], ["surgelé"], ["conserve"], 
            substr_3, ["alcoolisé"], ["non sucré"], substr_4]
# ajouter une colonne de "flag" utilisant les catégories de corpus
df["flag_cat"] = ''
foundWord = False

for cat in df["categories_fr"]:
    if(pd.isna(cat) == False):
        for corpus in corpus_cat: 
            for cor in corpus:
                if cor in str(cat):
                    index_mot = df[df["categories_fr"] == cat].index
                    df["flag_cat"].loc[index_mot] = corpus[0]
                    foundWord = True
                    break
            if foundWord:
                foundWord = False
                break
```
